```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import sys
sys.path.append('..')

import modules.plotting as pl
import modules.linear_algebra as lin
```

```python
%matplotlib inline
```

```python
points = np.array([[0, 0, 0], 
                   [1, 0, 0], 
                   [5, 4, 2], 
                   [1, 1, 0],
                   [8, 4, 3],
                   [-1, 2, 3],
                   [-5, -3, -5],
                  ])
```

```python
point, normal = lin.best_fit_plane(points)

_, direction = lin.best_fit_line(points)
```

#### Plane best fit

```python
fig, ax = plt.subplots()
ax = Axes3D(fig)

pl.scatter3(ax, points, color='k', depthshade=False)

# Calculate axis limits
x_min, x_max = ax.get_xlim3d()
x_range = np.arange(x_min, x_max)
y_min, y_max = ax.get_ylim3d()
y_range = np.arange(y_min, y_max)

pl.plot_plane(ax, point, normal, x_range=x_range, y_range=y_range, alpha=0.2)
```

#### Line best fit

```python
coeff = 10

line_point_1 = point - coeff * direction
line_point_2 = point + coeff * direction

x1, y1, z1 = line_point_1
x2, y2, z2 = line_point_2

fig, ax = plt.subplots()
ax = Axes3D(fig)

pl.scatter3(ax, points, color='k', depthshade=False)
ax.plot([x1, x2], [y1, y2], [z1, z2])
```
